```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mat
import sklearn.linear_model
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
import warnings as wr
import pickle
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/yash240990/Python/master/Startups_Data.csv")
data.head()
```

```python
data.info()
```

```python
data.columns
```

```python
#independend variable
x = ['R&D Spend', 'Administration', 'Marketing Spend', 'State']
x = data[x].values
x
```

```python
#depended variable
y = data.Profit.values
y
```

```python
#labelEncoder
from sklearn.preprocessing import LabelEncoder , OneHotEncoder
from sklearn.compose import ColumnTransformer

label_encoder = LabelEncoder()
x[:,3] = label_encoder.fit_transform(x[:,3])
transformer = ColumnTransformer(transformers=[("onehot",OneHotEncoder(),[3])],remainder='passthrough')
x = transformer.fit_transform(x)
x = x.astype(int)


```

```python
x
```

```python
#Split the dataset into Training amd Testing Subset
from sklearn.model_selection import train_test_split
#test_size=0.2 means 20% of data is send for testing and rest 80% for training
x_train , x_test ,y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print('Original Shape')
print(x.shape)
print(y.shape)
print('After splitting')
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(x_train,y_train)
```

```python
# pickle the model
pickle.dump(mlr,open('MLR_Model.pkl','wb'))
```

```python
pred_values = mlr.predict(x_test)
pred_values
```

```python
from sklearn.metrics import r2_score
acc = r2_score(y_test,pred_values)
print('Accuracy of model:',acc*100)
```
