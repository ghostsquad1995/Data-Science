```python
# unsupervised ML
'''
ðŸŸ¡ Apriori Algorithm

There are three major components of Apriori algorithm:

-Support:
Support(b) = (Transactions containing(b))/(Total Transactions)
Support(b) = ? %

-Confidence:
Confidence(Aâ†’B) = (Transactions containing both (A and B))/(Transactions containing A)
Confidence(Aâ†’B) = ? %
                        

-Lift:
Lift(Aâ†’B) = (Confidence (Aâ†’B))/(Support (B))
Lift(Aâ†’B) = ? %
'''
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
data = pd.read_csv(r"C:\Users\Prithviraj Ghorpade\Downloads\store_data.csv")
data
```

```python
records = []
for i in range(len(data)):
    records.append([str(data.values[i, j]) for j in range(len(data.columns))])
```

```python
association_rules = apriori(records, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)
```

```python
association_results = list(association_rules)
```

```python
# Print the association rules
for item in association_results:
    # Extract the rule, support, confidence, and lift
    pair = item[0]
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])
    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
